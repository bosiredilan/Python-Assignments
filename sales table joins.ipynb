```python
import pandas as pd
id_name_region=pd.read_csv(r"C:\Users\HP\Documents\Python_class\id_name_region.csv")
prod_description=pd.read_csv(r"C:\Users\HP\Documents\Python_class\prodid_prodname_category.csv")
customer_id_quantities=pd.read_csv(r"C:\Users\HP\Documents\Python_class\customer_id_quantities.csv")
region_manager=pd.read_csv(r"C:\Users\HP\Documents\Python_class\region_manager.csv")
category_description=pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\category_description.csv")

merged1=pd.merge(id_name_region,customer_id_quantities,on="CustomerID",how='inner')
merged1
```

```python
merged2=pd.merge(merged1,prod_description,on="ProductID",how='inner')
merged2
```

```python
merged3=pd.merge(merged2,region_manager,on="Region",how='left')
merged3
```
